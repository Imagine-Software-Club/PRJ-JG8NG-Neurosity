```python
index_prev = 0
index = 161

data_array = []
for i in range(0,df.shape[0]):
    row_array = np.median(df_raw_data.iloc[index_prev : index].values, axis=0)
    index_prev = index
    index += 161    
    print(index)
    data_array.append(row_array.tolist())
    
colnames = df_raw_data.columns.tolist()
df_combined = pd.DataFrame(data=data_array, columns=colnames)
df_combined
```

```python
#MIN MAX Scaling
def normalize_dataframe(df):
    return (df - df.min()) / (df.max() - df.min())
```

```python
from sklearn.decomposition import FastICA
```

```python
n_components = 8
ica = FastICA(n_components=n_components)
S_ = ica.fit_transform(df['probability'].reshape(-1, 1))  # Reconstruct signals
A_ = ica.mixing_  # Get estimated mixing matrix

# Convert to DataFrame
df_ica = pd.DataFrame(S_, columns=[f'IC_{i+1}' for i in range(n_components)])
```

```python
df_ica.plot()
```
